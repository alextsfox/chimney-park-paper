```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import seaborn as sns
```

```python
def parse_file(f):
    df = pd.read_excel(f)
    df["Survey_Date"] = pd.to_datetime(df["Survey_Date"], format=r"%Y%m%d")
    df["Survey_DOY"] = df["Survey_Date"].dt.dayofyear

    df["Stand_Name"].replace("BBNF", "BB-NF", inplace=True)
    df["Stand_Name"].replace("BBUF", "BB-UF", inplace=True)
    df["Stand_Name"].replace("BBSF", "BB-SF", inplace=True)
    df["Stand_Name"].replace("SF", "BB-SF", inplace=True)
    df["Stand_Name"].replace("UF", "BB-UF", inplace=True)
    df["Stand_Name"].replace("NF", "BB-NF", inplace=True)
    
    df["Stand_Name"] = np.where((df["Stand_Name"] == "EC") & (df["Plot_Number"].isin([3, 4, 5])) & (df["Survey_Year"] >= 2018), "BB-SF", df["Stand_Name"])
    df["Stand_Name"].replace("EC", "BB-NF", inplace=True)



    return df.loc[df["Stand_Name"].isin(["BB-UF", "BB-SF", "BB-NF"])]

seedsap = parse_file("CPk_VegSurvey_SeedSap_AllYears.xlsx")
# seedsap = (
#     seedsap[["Stand_Name", "Survey_Year", "Pine.Seedling.Count", "Aspen.Seedling.Count", "Pine.Sapling.Count", "Aspen.Sapling.Count"]]
#     .groupby(["Stand_Name", "Survey_Year"])
#     .mean()
#     .reset_index()
# )


trees = parse_file("CPk_VegSurvey_Trees_AllYears.xlsx")
trees["Condition"].replace("CFD", "D", inplace=True)
trees["Condition"].replace("BD", "D", inplace=True)
trees["Condition"].replace("FD", "D", inplace=True)
trees["Condition"].replace("Grey", "D", inplace=True)
trees["Condition"].replace("BG", "L", inplace=True)
trees["Condition"].replace("FG", "L", inplace=True)

trees["Basal_Area_m2"] = np.pi*trees["DBH"]**2/4*1/10_000
trees = trees[["Stand_Name", "Plot_Number", "Survey_Year", "Basal_Area_m2", "Condition"]].groupby(["Stand_Name", "Plot_Number", "Survey_Year", "Condition"]).sum().reset_index()
trees["Basal_Area_Index_m2_ha-1"] = trees["Basal_Area_m2"]/150*10000



vegcover = parse_file("CPk_VegSurvey_VegCover_AllYears.xlsx")
def cover_class_to_pct(coverclass):
    match coverclass:
        case 0: return 0
        case 1: return stats.uniform.rvs(0, 5)
        case 2: return stats.uniform.rvs(5, 20)
        case 3: return stats.uniform.rvs(25, 25)
        case 4: return stats.uniform.rvs(50, 25)
        case 5: return stats.uniform.rvs(75, 25)
        case _: return np.nan
for ctype in ["Burn", "Shrub", "Forb", "Graminoid", "Litter", "Lichen.or.moss", "Bare.soil.or.Rock"]:
    cover_pct = [np.nan]*vegcover.shape[0]
    for i in range(vegcover.shape[0]):
        cover_pct[i] = cover_class_to_pct(vegcover[ctype].iloc[i])
    vegcover[ctype] = cover_pct
vegcover["Total_Plant"] = vegcover["Shrub"] + vegcover["Forb"] + vegcover["Graminoid"] + vegcover["Lichen.or.moss"]
```

```python
fig, axs = plt.subplots(4, 2, figsize=(15,10))
axs = axs.flatten()

leg = True
for i, ctype in enumerate(["Burn", "Shrub", "Forb", "Graminoid", "Litter", "Lichen.or.moss", "Bare.soil.or.Rock", "Total_Plant"]):
    if i > 0: leg = False
    # sns.stripplot(x=vegcover["Survey_Year"], y=vegcover[ctype], hue_order=["BB-NF", "BB-UF", "BB-SF"], dodge=True, hue=vegcover["Stand_Name"], ax=axs[i], palette={"BB-NF":"C0", "BB-UF":"C1", "BB-SF":"C3"}, legend=leg, s=5, alpha=0.5)
    sns.violinplot(x=vegcover["Survey_Year"], y=vegcover[ctype], hue_order=["BB-NF", "BB-UF", "BB-SF"], dodge=True, hue=vegcover["Stand_Name"], ax=axs[i], palette={"BB-NF":"C0", "BB-UF":"C1", "BB-SF":"C3"}, cut=0, inner="stick", legend=False)

fig.suptitle("Ground Cover (%)")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs = axs.flatten()

leg = True
for i, dtype in enumerate(["L", "D"]):
    if i == 0: axs[i].set_title("Living")
    if i == 1: axs[i].set_title("Dead")
    sns.stripplot(x=trees.query("Condition == @dtype")["Survey_Year"], y=trees.query("Condition == @dtype")["Basal_Area_Index_m2_ha-1"], dodge=True, hue_order=["BB-NF", "BB-UF", "BB-SF"], hue=trees.query("Condition == @dtype")["Stand_Name"], ax=axs[i], palette={"BB-NF":"C0", "BB-UF":"C1", "BB-SF":"C3"})#, legend=leg)
fig.suptitle("DBH")
```

```python
fig, axs = plt.subplots(2, 1, figsize=(15, 5))
axs = axs.flatten()

leg = True
for i, ptype in enumerate(["Pine.Seedling.Count", "Aspen.Seedling.Count"]):
    if i > 0: leg = False
    sns.stripplot(x=seedsap["Survey_Year"], y=seedsap[ptype]/150, hue_order=["BB-NF", "BB-UF", "BB-SF"], hue=seedsap["Stand_Name"], ax=axs[i], legend=False, dodge=True, palette="dark:k")
    sns.boxplot(x=seedsap["Survey_Year"], y=seedsap[ptype]/150, hue_order=["BB-NF", "BB-UF", "BB-SF"], hue=seedsap["Stand_Name"], ax=axs[i], palette={"BB-NF":"C0", "BB-UF":"C1", "BB-SF":"C3"})
    # sns.violinplot(x=seedsap["Survey_Year"], y=seedsap[ptype]/150, hue_order=["BB-NF", "BB-UF", "BB-SF"], hue=seedsap["Stand_Name"], ax=axs[i], palette={"BB-NF":"C0", "BB-UF":"C1", "BB-SF":"C3"}, cut=0)
    # axs[i].set_yscale("log")
fig.suptitle("Seedling/Sapling Density (#/m2)")
```
